## **Paragraph**

```python
import spacy
nlp=spacy.load("en_core_web_sm")
import string
import nltk
from nltk.stem.porter import *
from nltk import word_tokenize
nltk.download('punkt')
p_stemmer = PorterStemmer()

# Paragraphs:
text_1=("""Pakistan: A land of diverse cultures, breathtaking landscapes from the towering peaks of the Himalayas to the vast deserts of Thar. Rich history, vibrant cities like Lahore and Karachi, and warm hospitality define its essence.
Australia: Vast continent-country, stunning landscapes, from the iconic Sydney Opera House to the awe-inspiring Uluru. Rich biodiversity, unique wildlife like kangaroos and koalas, and vibrant multicultural cities characterize its allure.
""")
```

## **Tokenization**


```python
def Tokenization(doc):
  doc=nlp(doc)
  token = [token.text for token in doc]
  print(token)
  return doc

tokenize_text=Tokenization(text_1)
```


## **StopWords Removal**

```python
def StopWords_Removal(doc):
    stopword_removal = [t.text for t in doc if not  t.is_stop]
    return stopword_removal

print("\nBefore Removing Stopwords:\n\n",tokenize_text, "\n")
Stopwords_removed=StopWords_Removal(tokenize_text)
print("\nAfter Removing Stopwords:\n\n",Stopwords_removed, "\n")
```

## **POS_Tagging**

```python
def POS_Tagging(doc):
    print(f'{"Text":{20}} {"POS":{20}} {"TAG":{20}} {"Is_alpha":{20}} {"Shape":{10}} ')
    for t in doc:
        print(f'{t.text:{20}} {t.pos_:{20}} {t.tag_:{23}} {t.is_alpha:{10}} {t.shape:{10}} ')

print("POS_Tagging:",end="\n\n")
POS_Tagging(tokenize_text)
```

## **Lemmatization**

```python
def Lemmatization(doc):
    print(f'{"Text":{20}} {"Lemma":{20}} {"Lemma Hash":{20}} ')
    for t in doc:
        print(f'{t.text:{20}} {t.lemma_:{20}} {t.lemma:{20}}')

print("\nLemmatization:",end="\n\n")
Lemmatization(tokenize_text)
```

## **Stemming**

```python
def Stemming(doc):
    for word in doc:
      print(f'{word:{20}} {"-->"} {p_stemmer.stem(word)}')

print("\nStemming:",end="\n\n")
words=word_tokenize(text_1)
Stemming(words)
```

## **Vocab Function's**


```python
def vocab(doc):
    print("Length of nlp.vocab:", len(nlp.vocab))
    print("nlp.vocab:", nlp.vocab)
    print("Length of doc.vocab:", len(doc.vocab))
    print("doc.vocab:", doc.vocab)
    for t in range(0,100):
      print(nlp.vocab[t].text,end=" | ")


vocab(tokenize_text)
```
