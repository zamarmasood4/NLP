```python
import nltk
from nltk import pos_tag, ne_chunk, word_tokenize
from IPython.display import display, HTML

text = '''Zamar Inc. is a dynamic technology firm specializing in cutting-edge of software solutions. With a commitment to excellence in Pakistan, we empower businesses to United at Pakistan. Our tailored approach is to fight with Thiland. So we will fight with India. Join Zamar Inc. as we embark on a journey to redefine possibilities in the realm of technology in Punjab, Pakistan.'''
pos_tag = pos_tag(word_tokenize(text))
txt = ne_chunk(pos_tag)
output = ''
for chunk in txt:
    if isinstance(chunk, nltk.Tree):
        tag = chunk.label()
        title = f"<span style='Font-Weight: bold; color: rgba(190,18,18,1);'> Named Entity Relations using NLTK: "
        output =output + f"<span style='padding:3px;border-radius:11px; border: 2px solid black; color: black; background: linear-gradient(90deg, rgba(150,18,18,1) 9%, rgba(190,110,110,1) 60%, rgba(96,14,14,1) 80%);'>{chunk[0][0]} ({tag})</span> "
    else:
        output =output + chunk[0] + ' '

display(HTML(title))
display(HTML(output))
```
