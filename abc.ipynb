```python
import os
import spacy
import string
from autocorrect import Speller
import pickle
nlp = spacy.load("en_core_web_sm")
spell = Speller()
path = 'D:\\University Data\\NLP\\Deep Learning\\Text_Files'
filepath = os.listdir(path)

D_F = []

def filetext(filename):
    for file in filename:
        text=open(path+"\\"+file,"r").readlines() 
        for sentence in text:
            D_F.append(sentence.strip())
filetext(filepath)

def correct_spelling(text):
    return spell(text)

def tokenize(text):
    return [token.text for token in nlp(text)]

def pos_tagging(text):
    return [token.text for token in nlp(' '.join(text))]

def remove_stopwords(text):
    return [token for token in text if token.isalpha() and token.lower() not in nlp.Defaults.stop_words and token not in string.punctuation]

def LLC(tokens):
    return [token.lower() for token in tokens]

def lemmatize(text):
    return [token.lemma_ for token in nlp(' '.join(text))]

def list_to_text(tokens):
    text = ' '.join(tokens)
    return text

D_S_P = []
for sentence in D_F:
    DSP_items = []
    sp = correct_spelling(sentence)
    DSP_items.append(sp)
    tokens = tokenize(sp)
    DSP_items.append(tokens)
    pos = pos_tagging(tokens) 
    DSP_items.append(pos)
    text_from_tokens = list_to_text(tokens)
    remove = remove_stopwords(pos) 
    DSP_items.append(remove)
    Lc = LLC(remove)
    DSP_items.append(Lc)
    lemma = lemmatize(Lc)
    DSP_items.append(lemma)

    D_S_P.append(DSP_items)


save=open("PreProcessed_Data.pickle","wb")
pickle.dump(D_S_P,save)
save.close()

BOW = []

for index in range(len(D_S_P)):
    lemmas = D_S_P[index][5]
    for l in lemmas:
        BOW.append(l)

print("The length of Bag of words is",len(BOW))
print(BOW)
```
