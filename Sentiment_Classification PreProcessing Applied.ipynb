```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
documents = []

for r in short_pos.split('\n'):
    documents.append( (r, "pos") )

for r in short_neg.split('\n'):
    documents.append( (r, "neg") )


print(documents[:5])
len(documents)
```

```python
import nltk
from nltk import word_tokenize
from autocorrect import Speller
import spacy

nltk.download('punkt')
nlp = spacy.load("en_core_web_sm")
spell = Speller()
all_words = []

def preprocessing(text):
    text1=spell(text)
    doc = nlp(text1)
    tokens = [token.text.lower() for token in doc if not (token.is_stop or not token.is_alpha)]
    lemmas=[token.lemma_ for token in nlp(' '.join(tokens))]
    return lemmas

short_pos_words = preprocessing(short_pos)
short_neg_words = preprocessing(short_neg)
print(len(short_pos_words))
print(len(short_neg_words))

for w in short_pos_words:
    all_words.append(w)
# print(len(all_words))
for w in short_neg_words:
    all_words.append(w)
# print(len(all_words))
all_words = nltk.FreqDist(all_words)
len(all_words)
```

```python
word_features = list(all_words.keys())[:5000]
word_features
```

```python
def find_features(document):
    words = word_tokenize(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)

    return features
```

```python
featuresets = [(find_features(rev), category) for (rev, category) in documents]
```

```python
import random

random.shuffle(featuresets)

featuresets[0]
```
